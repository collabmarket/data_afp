```python
# Archivos de valor cuota y valor patrimonio un fondo todas las AFP
fA = pd.read_csv('data/fA.csv', header=[0,1], index_col=0, parse_dates=True)
fB = pd.read_csv('data/fB.csv', header=[0,1], index_col=0, parse_dates=True)
fC = pd.read_csv('data/fC.csv', header=[0,1], index_col=0, parse_dates=True)
fD = pd.read_csv('data/fD.csv', header=[0,1], index_col=0, parse_dates=True)
fE = pd.read_csv('data/fE.csv', header=[0,1], index_col=0, parse_dates=True)

# Archivos de valor cuota un fondo todas las AFP
vcfA = pd.read_csv('data/vcfA.csv', index_col=0, parse_dates=True)
#vcfB = pd.read_csv('data/vcfB.csv', index_col=0, parse_dates=True)
vcfC = pd.read_csv('data/vcfC.csv', index_col=0, parse_dates=True)
#vcfD = pd.read_csv('data/vcfD.csv', index_col=0, parse_dates=True)
vcfE = pd.read_csv('data/vcfE.csv', index_col=0, parse_dates=True)
```

```python
fA.xs('Valor Cuota', axis=1, level=1).plot(figsize=[10,10], title='Fondo A')
#vcfA.plot(figsize=[10,10])
fB.xs('Valor Cuota', axis=1, level=1).plot(figsize=[10,10], title='Fondo B')
#vcfB.plot(figsize=[10,10])
fC.xs('Valor Cuota', axis=1, level=1).plot(figsize=[20,20], title='Fondo C')
#vcfC.plot(figsize=[20,20])
fD.xs('Valor Cuota', axis=1, level=1).plot(figsize=[10,10], title='Fondo D')
#vcfD.plot(figsize=[10,10])
fE.xs('Valor Cuota', axis=1, level=1).plot(figsize=[10,10], title='Fondo E')
#vcfE.plot(figsize=[10,10])
```

```python
fA.xs('Valor Patrimonio', axis=1, level=1).plot(figsize=[10,10], title='Valor Patrimonio Fondo A')
```

```python
fC['CUPRUM'].plot(figsize=[10,10], subplots=True, title='AFP CUPRUM Fondo C')
```

```python
vcfC.loc[:,('CUPRUM','HABITAT','MODELO')].resample('A').mean().pct_change(1).plot(
    title='Rentabilidad promedio anual Fondo C')
## Observaciones
# Se puede ver que la rentabilidad promedio anual va disminuyendo
# Las reltabilidad promedior anual entre distintas AFP no varían mucho
```

```python
cuprum_vc = pd.read_csv('data/VC-CUPRUM.csv', index_col=0, parse_dates=True)
cuprum_vc.loc[:,('A','C','E')].resample('A').mean().pct_change(1).plot(
            title='AFP CUPRUM rentabilidad promedio anual Fondo A,C,E')
## Observaciones
# Se puede ver que la rentabilidad promedio anual va disminuyendo
# Las reltabilidad promedior anual entre distintos Fondos. 
```
